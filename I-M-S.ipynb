```python
#Instructions
print("Welcome to IMS")
print()
print("A-Add an Product")
print("R-Remove an Product")
print("E-Edit specifics of an Products")
print("L-List all the Products")
print("I-Inquire about a product ")
print("P-Purchase a product")
print("C-Checkout")
print("S-Show all products purchased")
print("Q-Quit")
print("remove-Remove an product from the cart")
print("help-See all commands again")
print()

import random
import time
import datetime

#List to store product ID's the product purchased
cart=[]

c="y" #Runs the while loop as long as user wants

pn = {}
tq = 0

total_cost=0 
flag=0 #To check if they have checked out

user = input('Enter Your Name: ')

print()
print('Welcome ', user, '!!')

while(c!= "q" or c!= "Q"):
    print()
    c= input("What would you like to do? ")
    
    if(c=="q" or c=="Q"):
        break
        
    
    #Add an product-----
    
    elif(c=="A" or c=="a"):
        prod_id = (input("Enter product number: "))
        
        if prod_id in record.keys():
            print()
            print("Product ID already exist!!")
            print("Are you sure, you wanna edit the Product?")
            print()
            d = input("If yes enter 'y' else 'n':")
            if d == 'Y' or d =='y':
                print()
                
            elif d == 'n' or d == 'N':
                prod_id = input("Enter product number: ")
                if prod_id in record.keys():
                    print()
                    print("This Product ID also already exist!!")
                    print("So, please check the already existing Product IDs and Enter you choices again carefully.")
                    print("If you wanna edit choose 'e' to edit a product")
                    print()
                    continue

            else:
                print()
                print("Please enter in valid character!!")
                print('Enter the choices again carefully')
                print()
                continue
                
        i_cat = input("Enter product category: ")
        i_name = input("Enter product name: ")
        i_pr = float(input("Enter product price: Rs."))
        i_desc = input("Enter product description: ")
        i_quant = int(input("Enter product Quantity: "))
        
        if i_quant < 0:
            print()
            print("Enter the correct details, as qunatity cannot be less than zero.")
            print('Please be careful with choices.')
            print("Enter 'a' again to add the product!!")
            continue
               
        record[prod_id] = {'Category': i_cat,'Prod_name': i_name, 'Description': i_desc, 'Quantity': i_quant, "Price Rs.": i_pr}
        
        print()
        
        print('{','Product ID: ', prod_id,", ", "Category: ", i_cat, ", " , "Product Name: ", i_name,", " , "Description: ", i_desc,", " , "Qunatity: ", i_quant,", " , "Price: Rs.",i_pr,"}")
        print('Item was successfully added.')
        
    
    #Edit a product------
    
    elif(c=="E" or c=="e"):
        print()
        prod_id = input("Enter product ID: ")
        
        if prod_id in record.keys():
            i_name = input('Enter the product name: ')
            i_pr = float(input("Enter product price: "))
            i_desc = input("Enter product description: ")
            i_quant = int(input("Enter product quantity: "))
            
            if i_quant < 0:
                print()
                print('The quantity of product cannot be negative.')
                q = input("Enter 'c' to rewrite the quantity, else quantity will be set to 0.")
                if q == 'c' or q == 'C':
                    i_quant = int(input("Enter product quantity: "))
                    if i_quant < 0:
                        print()
                        print("Quantity cannot be Negative!!")
                        print('Setting the Quantity to 0.')
                        i_quant = 0
                        print('Quantity: ', i_quant)
                        print()
                    
                else:
                    i_quant = 0
                    print('Qunatity of Product is now set to 0.')
                    print('Quantity: ', i_quant)
                    print()
                
            
            record[prod_id] = {'Category': i_cat,'Prod_name': i_name, 'Description': i_desc, 'Quantity': i_quant, "Price Rs.": i_pr}
        
        else:
            print("That item does not exist, to add an item use 'a'")
        print()
        
   

    #Remove a product------
    
    elif(c=="R" or c=="r"):
        print()
        prod_id = input("Enter Product ID: ")
        if(prod_id in record.keys()):
            are_you_sure = input("Are you sure you want to remove that item(y/n)? ")
            if(are_you_sure=="y" or are_you_sure=="Y"):
                p = record.pop(prod_id)
                
                
                print('Product', p, 'deleted successfully.')
                print()
                print("Item successfully removed!")
            print()
        else:
            print("Sorry, we don't have such an product!")
            print()
            
    
    #List all the products------
    
    elif(c=="L" or c=="l"):
        
        s = set()
        for i in record.keys():
            s.add(record[i]['Category'])
        cate = list(s)
        
        print()
        print('All the valid Product IDs: ', list(record.keys()))
        print()
        print("All product categories: ", list(s) )
        print()
        print("Product Menu-- ")
        print()

        for j in range(len(cate)):
            print('\t','---------------------------------------')
            print('\t', '      | ',cate[j], ' |')
            print('\t','---------------------------------------')
            for i in record.keys():
                if record[i]['Category'] == cate[j]:
                    print('Prod ID: ', i ,',  ','Product Name: ', record[i]['Prod_name'], ',  ' , 'Price: Rs.', record[i]['Price Rs.'], ',  ', 'Quantity: ', record[i]['Quantity'] )
    
    
    
    #Inquire about a product------
    
    elif(c=="I" or c=="i"):
        print()
        prod_id = input("Enter Product IDs: ")
        if(prod_id in record.keys()):
            print()
            print("Product ID: ",prod_id ,', ','Product Name: ',record[prod_id]['Prod_name'], ',', " Description: ", record[prod_id]['Description'],', '," Price: ",record[prod_id]['Price Rs.'],', '," Quantity: ",record[prod_id]['Quantity'])
            if(record[prod_id]['Quantity'] < 3 and record[prod_id]['Quantity'] != 0):
                print()
                print("*****Only ",record[prod_id]['Quantity']," remaining! Hurry!*****")
            print()
        else:
            print("Sorry we don't have such an item!")
            print()  
            
    
    
    #Purchase a product------
    
    elif(c=="P" or c=="p"):
        print()
        prod_id  = input("Enter the product_Id: ")
    
        if(prod_id in record.keys()):
            o_quant = int(input("Enter the quantity: "))
            if o_quant < 0:
                print()
                print('You cannot order 0 quantity!!')
                order = input("Enter 'y' to continue ordering:")
                if order == 'y' or order == 'Y':
                    o_quant = int(input("Enter the quantity: "))
                    print()
                    if o_quant < 0:
                        print('You cannot order 0 quantity!!')
                        print("Enter 'p' to place the order  again. ")
                        print()
                        continue
                    else:
                        print()

                else:
                    print()
                    print("Enter 'p' to place the order  again. ")
                    continue


            if( flag == 1):
                flag = 0
            stock = record[prod_id]['Quantity']
            print()
            if stock == 0:
                print()
                print("Sorry !! We don't have product in stock.")
            elif stock < o_quant:
                print()
                print("Sorry!! We only have ", stock, 'stocks left!!')
                print()
                print('You can order', stock, 'Stock or you can select anything else from the list.')
                print("To see the List enter 'L'." )
            else:
                item_price = record[prod_id]['Price Rs.'] * o_quant
                print("Product Name: ", record[prod_id]['Prod_name'])
                print("Price : ", record[prod_id]['Price Rs.'])
                total_cost = total_cost + item_price
                record[prod_id]['Quantity'] = record[prod_id]['Quantity'] - o_quant
                print('Product ', record[prod_id]['Prod_name'],' with ', o_quant,'quantity ', 'added to cart:', 'Rs.', item_price)
                cart.append(prod_id) #stores item in cart
                pn[record[prod_id]['Prod_name']] = {'Qty':o_quant, 'MRP': record[prod_id]['Price Rs.'], 'GST': 18, 'Amount': item_price}
                print()
                print("Enter 'P' to purchase again!!")
                

        else:
            print()
            print('Product ID does not exist!!')
            print("Please enter valid Product ID, enter 'l' to know about all.")
            print()
            
    
    
    #Check out------
    
    elif(c=="C" or c=="c"):
        print()
        print("You bought the following products: ",cart)
        d = str(datetime.datetime.now().year) + '-' + str(datetime.datetime.now().month) + '-' + str(datetime.datetime.now().day)
        dt = str(d)
        t = str(datetime.datetime.now().hour) + ':' + str(datetime.datetime.now().minute) + ':' + str(datetime.datetime.now().second)
        ti = str(t)
        for k in pn.keys():
            print()
            print('Product Bought:', k, ', ','Quantity Bought:', pn[k]['Qty'])
        print()
        print('\t','---------------------------------------')
        print('\t','\t', 'Inventory Management System')
        print('\t','---------------------------------------')
        print('\t', "Order No. Take Out", '\t', 'Tax Invoice')
        print('------------------------------------------------------------------------------------------')
        print('Bill No.: ', round((random.uniform(1,10))*1000), '\t', 'Bill: ' ,'Cash Bill')
        print('Counter: ', 'CASH1', '\t', 'Date: ', dt)
        print('User: ', user, '\t', '  ', 'Time: ', ti)
        print('------------------------------------------------------------------------------------------')
        print('Product Name','        |   ','\t', '   ','Qty','   |   ', 'MRP','   |   ', 'GST%','  |  ', 'Amount','  |')
        print('------------------------------------------------------------------------------------------')
        for v in pn.keys():
            print()
            print(v, '    |    ', pn[v]['Qty'],'    |   ', pn[v]['MRP'],'    |    ', pn[v]['GST'], '    |    ',pn[v]['Amount'] )
            tq += pn[k]['Qty']
        print()
        print('------------------------------------------------------------------------------------------')
        
        print('Total Qty: ',tq , '              ',"Gross: ","Rs.",round(total_cost,2))
        tax= round(0.18*total_cost,2)
        print('Total Products: ', len(pn.keys()),'               ' ,"Tax is 18%: ","Rs.",tax)
        total = round(total_cost+tax,2)
        print()
        print('------------------------------------------------------------------------------------------')
        print("Total including Tax: ",'   ' ,"Rs.",total)
        print('------------------------------------------------------------------------------------------')
        total_cost=0
        flag=1
        print()
        print("You can still purchase items after check out, your cart has been reset. To quit press q")
        print()
        
    
    
    #Display all commands------
    
    elif(c=="help" or c == "HELP"):
        
        print()
        print("A-Add an Product")
        print("R-Remove an Product")
        print("E-Edit specifics of an Products")
        print("L-List all the Products")
        print("I-Inquire about a product ")
        print("P-Purchase a product")
        print("C-Checkout")
        print("S-Show all products purchased")
        print("Q-Quit")
        print("remove-Remove an product from the cart")
        print("help-See all commands again")
        print()
        
    
    #To remove an product from the cart-----
    
    elif(c=="remove" or c=="Remove"):
        print()
        print('It will remove only 1 product quantity at a time')
        print("You can quit by 'q' and reset the cart also by entering 'n'.")
        print()
        are_you_sure = input("Are you sure you want to remove an Product from the cart(y/n)? ")
        
        if(are_you_sure=="y"):
            print()
            prod_id  = input("Enter the product_Id: ")

            if(prod_id in cart):
                
                print()
                item_price = record[prod_id]['Price Rs.'] * 1
                print("Product Name: ", record[prod_id]['Prod_name'])
                print("Price : ", record[prod_id]['Price Rs.'])
                total_cost = total_cost - item_price
                record[prod_id]['Quantity'] = record[prod_id]['Quantity'] + 1
                
                j=0
                for i in range(0,len(cart)):#To find the index of the product in the list cart
                    if(i==prod_id):
                        j=i

                cart.pop(j)
                
                print('Product ', record[prod_id]['Prod_name'],' with ', '1','quantity ', 'removed from cart having price :', 'Rs.', item_price)
                print()
            else:
                print()
                print("That item is not in your cart!")
                print()
                
    
    
    #prints list cart------
    
    elif(c=="s" or c=="S"):
        print()
        print(cart)
        print('-------------------------------------------------------------------------------------')
        for k in pn.keys():            
            print('Product Bought:', k, ', ','Quantity Bought:', pn[k]['Qty'])
        print('-------------------------------------------------------------------------------------')
        print()
        
        
    else:
        print()
        print("ERROR! Please make choices correctly!")
        print("Enter 'help' to know all the existing commands.")
        print()
        
        
        
#Outputs total if the user quits without checking out-----

if(total_cost>0 and flag==0):
    print()
    d = str(datetime.datetime.now().year) + '-' + str(datetime.datetime.now().month) + '-' + str(datetime.datetime.now().day)
    dt = str(d)
    t = str(datetime.datetime.now().hour) + ':' + str(datetime.datetime.now().minute) + ':' + str(datetime.datetime.now().second)
    ti = str(t)
    print('\t','-------------------------------------------')
    print('\t','\t', 'Inventory Management System')
    print('\t','-------------------------------------------')
    print("Order No. Take Out", '\t','\t', 'Tax Invoice')
    print('------------------------------------------------------------------------------------')
    print('Bill No.: ', round((random.uniform(1,10))*1000), '\t', 'Bill: ' ,'Cash Bill')
    print('Counter: ', 'CASH1', '\t', 'Date: ', dt)
    print('User: ', user, '\t','  ', 'Time: ', ti)
    print('------------------------------------------------------------------------------------')
    print('Product Name','\t','            | ', '  ','Qty', '   ','  |  ', 'MRP','   ','  |  ', 'GST%', '   ','  |  ', 'Amount','  |  ')
    print('------------------------------------------------------------------------------------')
    for v in pn.keys():
        print()
        print(v, '     ','   |   ', pn[v]['Qty'],'   |   ', pn[v]['MRP'],'   |   ', pn[v]['GST'],'   |   ',pn[v]['Amount'],'   |  ')
        tq += pn[v]['Qty']
    print()
    print('----------------------------------------------------------------------------')
        
    print('Total Qty: ',tq , '         ',"Gross: ","Rs.",round(total_cost,2))
    tax= round(0.18*total_cost,2)
    print('Total Products: ', len(pn.keys()) ,'        ',"Tax is 18%: ","Rs.",tax)
    total = round(total_cost+tax,2)
    print()
    print('----------------------------------------------------------------------------')
    print("Total including Tax: ",'   ' ,"Rs.",total)
    print('----------------------------------------------------------------------------')
    
print()
print('\t',"Thank you for using IMS")
print('\t','See you Again :)')


js = json.dumps(record)

fd = open("records1.json",'w')
fd.write(js)
fd.close()
```
